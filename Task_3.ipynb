```python
plt.figure(figsize=(8,8))
sns.countplot(data=dataset,x='type',palette='Set1')
plt.xlabel('catogories',fontsize=10)
plt.ylabel('Counts',fontsize=10)
plt.title('movies vs tvshows',fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show
```

```python
plt.figure(figsize=(8,6))
sns.countplot(data=dataset,x='rating',order=dataset['rating'].value_counts().index,palette='Set2')
plt.xlabel('Rating',fontsize=10)
plt.ylabel('Counts',fontsize=10)
plt.title('Rating distribution',fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show
```

```python
plt.figure(figsize=(8,8))
sns.histplot(data=dataset,x='release_year',bins=30,kde=True,color='red')
plt.xlabel('year of release',fontsize=10)
plt.ylabel('Counts',fontsize=10)
plt.title('Release year distribution',fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show
```

```python
plt.figure(figsize=(8,8))
top_countries=dataset['country'].value_counts().head(10)
sns.barplot(x=top_countries.values,y=top_countries.index,palette='muted')
plt.xlabel('count',fontsize=10)
plt.ylabel('Country',fontsize=10)
plt.title('countries with most shows',fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show
```

```python
plt.figure(figsize=(8,8))
genres=dataset['listed_in'].str.split(',').explode().str.strip()
top_genres=genres.value_counts().head(10)
sns.barplot(x=top_genres.values,y=top_genres.index,palette='dark:red')
plt.xlabel('count',fontsize=10)
plt.ylabel('Genre',fontsize=10)
plt.title('top genres',fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show
```
